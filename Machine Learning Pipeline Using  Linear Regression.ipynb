```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('data_for_ml.csv')
df.head()
```

```python
# Check missing values
print(df.isnull().sum())

# Fill numerical columns with mean
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Fill categorical columns with mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Remove duplicates
df = df.drop_duplicates()

df.info()
```

```python
le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col])

df.head()
```

```python
y = df['Final_Score']
```

```python
X1 = df[['Hours_Studied']]

model1 = LinearRegression()
model1.fit(X1, y)

y_pred1 = model1.predict(X1)

print("Model 1 - Hours_Studied")
print("MSE:", mean_squared_error(y, y_pred1))
print("R2 Score:", r2_score(y, y_pred1))
```

```python
X2 = df[['Sleep_Hours']]

model2 = LinearRegression()
model2.fit(X2, y)

y_pred2 = model2.predict(X2)

print("Model 2 - Sleep_Hours")
print("MSE:", mean_squared_error(y, y_pred2))
print("R2 Score:", r2_score(y, y_pred2))
```

```python
X3 = df[['Internet_Usage']]

model3 = LinearRegression()
model3.fit(X3, y)

y_pred3 = model3.predict(X3)

print("Model 3 - Internet_Usage")
print("MSE:", mean_squared_error(y, y_pred3))
print("R2 Score:", r2_score(y, y_pred3))
```

```python
X4 = df[['Attendance']]

model4 = LinearRegression()
model4.fit(X4, y)

y_pred4 = model4.predict(X4)

print("Model 4 - Attendance")
print("MSE:", mean_squared_error(y, y_pred4))
print("R2 Score:", r2_score(y, y_pred4))
```

```python
X5 = df[['Gender']]

model5 = LinearRegression()
model5.fit(X5, y)

y_pred5 = model5.predict(X5)

print("Model 5 - Gender")
print("MSE:", mean_squared_error(y, y_pred5))
print("R2 Score:", r2_score(y, y_pred5))
```

```python
results = pd.DataFrame({
    'Feature': ['Hours_Studied', 'Sleep_Hours', 'Internet_Usage', 'Attendance', 'Gender'],
    'MSE': [
        mean_squared_error(y, y_pred1),
        mean_squared_error(y, y_pred2),
        mean_squared_error(y, y_pred3),
        mean_squared_error(y, y_pred4),
        mean_squared_error(y, y_pred5)
    ],
    'R2 Score': [
        r2_score(y, y_pred1),
        r2_score(y, y_pred2),
        r2_score(y, y_pred3),
        r2_score(y, y_pred4),
        r2_score(y, y_pred5)
    ]
})

results
```
